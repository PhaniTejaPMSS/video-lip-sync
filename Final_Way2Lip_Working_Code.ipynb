# **Key points to note :**
1. The input video file must not have any audio track and must be only a ".mp4" file.
2. The input audio file must be a ".wav" file and not ".mp3".
3. Both files should be of the exact same length.
4. *( Imp )* Target face in the input video , must be "detectable" in **ALL** videoframes. (So, no black or blurry frames, etc.)
5. This model is not reliable for very long and high res clips. (1080p / 30 seconds max)
6. The input files must be named as " *test_vid.mp4* " and " *test_aud.WAV* "
